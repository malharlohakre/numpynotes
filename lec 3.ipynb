```python
import numpy as np
```

```python
#zeros(shape,dtype=float/jiska chye uska ,order='c')
data = np.zeros((3,3),dtype =int)
print(data)
print(data.ndim)
data1 = np.zeros((3,3,2),dtype =int)
print(data1)
print(data1.ndim)
```

```python
#ones(shape,dtype=float/jo chye wo)
data = np.ones((3,2),dtype=int)
print(data)
```

```python
data = np.ones((2),dtype=int)
print(data)
data = np.ones((0),dtype=int)
print(data)
print(data.ndim)
data = np.ones((0,0),dtype=int)
print(data)
print(data.ndim)
```

```python
data = np.empty((2,2))
print(data)
```

```python
arr = np.full((3,3),np.pi)
print(arr)
```

```python
arr = np.full((3,3),'Hello',dtype='U5')# U defines how maany characters of string you want
print(arr)
```

```python
arr = np.full((3,3),'Hello',dtype='U3')
print(arr)
```

```python
data = np.identity(4 , dtype=int)
print(data)
```

```python
arr = np.arange(12)
re_arr = np.reshape(arr,(2,6),order="F")
print(arr)
print(re_arr)
print(re_arr.ndim)
```

```python
arr = np.arange(24)
re_arr = np.reshape(arr,(12,2))
print(arr)
print(re_arr)
print(re_arr.ndim)
```

```python
arr = np.arange(24).reshape(2,3,4)
print(arr)
print(arr.ndim)
```

```python
new_data = np.swapaxes(arr,0,2)
print(new_data)
```

```python
a1 = np.array([[12,13,14],[15,16,17],[18,19,20]])
a2 = np.array([[2,3,4],[5,6,7]])
concat = np.concatenate((a1,a2))
print(concat)
```
