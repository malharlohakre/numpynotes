```python
#series
data = ["java","python","c++","ruby","r"]
for i in range (len(data)):
    print(f'{i} {data[i]}')
```

```python
import pandas as pd
s = pd.Series(data)
print(s)

indexx  = ['a','b','b','c','d']
s = pd.Series(data,index = indexx)
print(s)
```

```python
import pandas as pd
data = [21,22,23,24,26,27]
s = pd.Series(data,dtype = float)
print(s)
```

```python
import numpy as np
data = [23,24,25,26,np.nan,28,29]
data_value = data[4]
print(type(data_value))# checking the data type of data here we checked nan
print(data_value)     
s = pd.Series(data,dtype = float)
print(s)
#adding null value
```
