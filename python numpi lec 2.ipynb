```python
import numpy as np
```

```python
#arange(start,stop,step,datatype) its abut genrating array
data = np.arange(0,10,2,dtype= int)
print(data)
```

```python
data = np.arange(0.5,5.5,1.0,dtype= float)
print(data)
```

```python
data = np.arange(10,2,-2,dtype= int)
print(data)
```

```python
#np.linspace(start,stop,num=50,endpoint=true,retstep=false,dtype 50 is default number set endpt & restep is also default)
data , step= np.linspace(0,10,5,endpoint = True,retstep = True)
print(data)
print("step size :",step)
```

```python
data , step= np.linspace(0,10,5,endpoint = False,retstep = True)
print(data)
print("step size :",step)
```

```python
data = np.logspace(1,3,4,base =10)
print(data)
```
