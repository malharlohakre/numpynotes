```python
#merging 
import pandas as pd
student = pd.DataFrame(
    {
        'studentID':[101,102,103,104,105],
        'name':['malhar','meet','om','ron','pri'],
        'age':[21,5,21,23,21]
    }
)
score = pd.DataFrame(
    {
        'studentID':[101,102,103,106],
        'subject':['math','math','science','history'],
        'Score':[90,99,82,83]
    }
)

print(student)
print(score)
```

```python
#inner join
# using merge
inner_merge = pd.merge(student,score,on='studentID')
print(inner_merge)
```

```python
#left join 
left_merge = pd.merge(student,score,on='studentID',how='left')
print(left_merge)
```

```python
#right join 
right_merge = pd.merge(student,score,on='studentID',how='right')
print(right_merge)
```

```python
#full/outer join 
full_merge = pd.merge(student,score,on='studentID',how='outer')
print(full_merge)
```
