```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# load data set
tips = sns.load_dataset('tips')
print(tips.head())
```

```python
tips.info()
```

```python
tips.describe()
```

```python
# scatter plot
sns.scatterplot(data=tips, x ='total_bill', y ='tip', hue='time')
#set title
plt.title('Total_bill vs Tip')
#set label
plt.xlabel('Total__Bill')
plt.ylabel('Tips')
plt.show()
```

```python
#line plot
sns.lineplot(data=tips, x ='total_bill', y ='tip', hue='time')
#set title
plt.title('Total_bill vs Tip')
#set label
plt.xlabel('Total__Bill')
plt.ylabel('Tips')
plt.show()
```

```python
#print column name
print(tips.columns)
```

```python
#loading another data set
fmri = sns.load_dataset('fmri')
print(fmri.head())
print(fmri.columns)
```

```python
sns.lineplot(data= fmri, x ='timepoint', y ='signal', hue='event' ,style='region')
#set title
plt.title('FMRI signal over time')
#set label
plt.xlabel('Timepoint')
plt.ylabel('Signal')
plt.show()
```

```python
# load data set
print(tips.head())

sns.barplot(data=tips,x='day',y='total_bill',hue='time')
plt.title('Average total bill per day by time')
plt.xlabel('Day')
plt.ylabel('Average total bill')
plt.savefig('barplot.png')# to save the figure 
plt.show()
```

```python
#histogram of total bill is histogram is made for distributions
sns.histplot(data=tips, x='total_bill', bins=10, kde=True)# kde kernel density estimation (add a density curve)
plt.title('Distribution of total bill')
plt.xlabel('Total bill')
plt.ylabel('Frequency')
plt.show()

```

```python
#box plot sumarize data distribution with respect to categories 
#highlight outliers
sns.boxplot(data=tips, x='day', y='total_bill', hue='smoker')
plt.title('Boxplot of total bill by day and smoker status')
plt.xlabel('day')
plt.ylabel('total_bill')
plt.show()
```

```python
# load iris dataset
iris = sns.load_dataset('iris')
print(iris.head())
```

```python
sns.pairplot(iris, hue='species')
plt.suptitle('Iris PAirplot', y=1.02)
plt.show()
```

```python
# violinplot plot it shows distribution across different
violinplot = sns.violinplot(data=iris, x='species', y='sepal_length', inner='quartile')
plt.title('Violin Plot of sepal length by species')
plt.xlabel('species')
plt.ylabel('sepal length')
plt.show()
```

```python
#countplot - count of each species
countplot= sns.countplot(data=iris, x='species')
plt.title('Count of each Irsi speices')
plt.xlabel('species')
plt.ylabel('count')
plt.show()
```

```python
#facet grid multiple grids based on categories
g = sns.FacetGrid(tips, col='time', row='smoker')
g.map(sns.scatterplot, 'total_bill', 'tip')
g.add_legend()
plt.show()
```

```python
# joint plot combination of scatter plot and histogram
sns.jointplot(data=tips, x='total_bill', y='tip', kind='scatter', hue='time')
plt.suptitle('Joint Plot of Total Bill and Tip', y=1.02)
plt.show()
```
