```python
#data frames 
#dictionary means it  adds a key and values
import pandas as pd
data = {
    'Name':['Rahul','Rohan','Tiya','Riya','Diana','Joy','Nishant','Heena','Deep'],
    'Age' :[23,34,45,67,24,12,50,22,56],
    'City':['Tokyo','Paris','Mumbai','Surat','Bhuj','Jaipur','Pune','Pune','Mumbai']
}
df = pd.DataFrame(data)
print(df)
```

```python
# conditional selection
#print row where amount > 3000
print(data[data['Amount'] > 3000])
```

```python
#print row amount > 4000 with transaction type online
print(data[(data['Amount'] > 4000) & (data['Transaction_Type'] == 'Online')]) # use parenthesis 
```

```python
print(data[(data['Amount'] > 2000) & (data['Category'] == 'Clothing') & (data['Transaction_Type'] == 'Online')])
```

```python
# export new_record to csv 
new_record = data[(data['Amount'] > 4000) & (data['Transaction_Type'] == 'Online')]
new_record.to_csv('new_record.csv',index = False)
```
