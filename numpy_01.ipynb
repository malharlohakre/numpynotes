```python
import numpy as np
```

```python
# array a collection of data & can handle large numericals data 
data = np.array([21,22,23,24,25])
print(data)
print(data[3])
print(type(data))
print(data.shape)
print(data.size)
print(data.ndim)
```

```python
data = np.array([[21,22,23],[24,25,26],[27,28,29]])
print(data.shape)
print(data)
print(data[1,1])
print(data.size)
print(data.ndim)
#2D array
```

```python
data = np.array([[[21,22,23],[24,25,26],[27,28,29]],[[21,22,23],[24,25,26],[27,28,29]]])
print(data.shape)
print(data.size)
print(data.ndim)
#3D array
```

```python
data = np.array([[[[21,22,23],[24,25,26],[27,28,29]],
                 [[21,22,23],[24,25,26],[27,28,29]]],
               [[[21,22,23],[24,25,26],[27,28,29]],
                 [[21,22,23],[24,25,26],[27,28,29]]]])
print(data.shape)
print(data.size)
print(data.ndim)
#4D array
```
