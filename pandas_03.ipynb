```python
#adding and modifying data 
import pandas as pd
data = {
    'Name':['Rahul','Rohan','Tiya','Riya','Diana','Joy','Nishant','Heena','Deep'],
    'Age' :[23,34,45,67,24,12,50,22,56],
    'City':['Tokyo','Paris','Mumbai','Surat','Bhuj','Jaipur','Pune','Pune','Mumbai']
}
df = pd.DataFrame(data)
print(df)
```

```python
# add new column
df['Salary'] = [5000,6000,7000,9000,6000,8000,4440,6398,1545]
print(df)
```

```python
#modifying existing column
df['Age'] = df['Age'] +2
print(df)
```

```python
# add new row
new_row = {'Name':'malhar','Age':21,'City':'Tokyo','Salary':5000}
#concat new row to the dataframe
df = pd.concat([df, pd.DataFrame([new_row])], ignore_index=True)
print(df)
```

```python
# another way
new_row = {'Name':'malhar','Age':21,'City':'Tokyo','Salary':5000}
row = pd.DataFrame([new_row])
#concat new row to the dataframe
df = pd.concat([df, row], ignore_index=True)
print(df)
```

```python
# adding none value
#import numpi to add none in numbers value like age salary
import numpy as np
new_row = {'Name':None,'Age': np.nan,'City':'Tokyo','Salary':5000}
row = pd.DataFrame([new_row])
#concat new row to the dataframe
df = pd.concat([df, row], ignore_index=True)
print(df)
```

```python
# hanling missing data
df.isnull().sum()
```

```python
# delete row null value
df = df.dropna()
print(df)
```

```python
# checking data gaya ya nahi
df.isnull().sum()
```

```python
new_row = {'Name':'Meet','Age': np.nan,'City':'Tokyo','Salary':np.nan}
row = pd.DataFrame([new_row])
#concat new row to the dataframe
df = pd.concat([df, row], ignore_index=True)
print(df)
```

```python
df.isnull().sum()
```

```python
#adding values inplace of nan
df['Age'].fillna(df['Age'].mean()) 
```

```python
df['Salary'] = df['Salary'].fillna(1)
```

```python
# add name column with Unknown where null
df['Name'] = df['Name'].fillna('Unknown')
print(df)
```

```python
# removing dupilcates value
#drop_duplicates is a inbuilt fuction
df= df.drop_duplicates()
print(df)
```

```python
#find the duplicates
duplicates = df.duplicated()
print(duplicates)
#find total duplicates 
total_duplicates = df.duplicated().sum()
print("Total duplicates ", total_duplicates)
```

```python
# it removes one duplicate
df = df.drop_duplicates(subset=['Name'])
print(df)
```

```python
# it removes  duplicate
df = df.drop_duplicates(subset=['City'])
print(df)
```

```python
# reset index 
#if after removing duplicates if the index value chnges use this
df = df.drop_duplicates().reset_index(drop=True)
print(df)
```

```python
#sort the data frame
#sort by age ascending
df = df.sort_values(by='Age')
print(df)
```

```python
# delete row null value
df = df.dropna()
print(df)
```

```python
# srot by descending
df = df.sort_values(by='Age', ascending = False)
print(df)
```

```python
# sort multiple columns
df = df.sort_values(by=['City','Age'])
print(df)
```

```python
# adding new rows 
new_row = {'Name':'malhar','Age': 22,'City':'Pune','Salary':5522}
row = pd.DataFrame([new_row])
#concat new row to the dataframe
df = pd.concat([df, row], ignore_index=True)
print(df)
```

```python
# delete row null value
df = df.dropna()
print(df)
```
